```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #to display wordcloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator 
from PIL import Image # to load our image
from os import path
```

```python
def upload_data():
    bookstore_df = pd.read_csv('data/cleaning_round2_blackwell.csv')
    return bookstore_df
```

```python
bookstore_df = upload_data()
```

```python
bookstore_df
```

```python
#dataframe['column_name'] = dataframe['column_name'].fillna('').apply(str)
```

```python
# Convert to string
bookstore_df['blurb_review']= bookstore_df['blurb_review'].fillna('').apply(str)
```

```python
bookstore_df.blurb_review.head(50)
```

```python
text = "".join(review for review in bookstore_df.blurb_review.astype(str))
```

```python
print ("There are {} words in the combination of all cells in column blurb_review.".format(len(text)))
```

```python
stopwords = set(STOPWORDS)
```

```python
stopwords.update(["br","hellip","s","p","em","b"])
```

```python
WordCloud().generate(text)
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=900, height=400).generate(text)
```

```python
plt.axis("off")
plt.figure( figsize=(60,20))
plt.tight_layout(pad=0)
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```

```python
book_mask = np.array(Image.open("9TRLzgqyc.jpg"))
book_mask
```

```python
text = "".join(review for review in bookstore_df.blurb_review.astype(str))
stopwords = set(STOPWORDS)
stopwords.update(["br","hellip","s","p","em","b"])

# Create a word cloud image

wc = WordCloud(background_color="white", max_words=100, mask=book_mask, stopwords=stopwords, contour_width=3, contour_color='green')

# Generate a wordcloud

wc.generate(text)

# store to file

wc.to_file("book_review.png")

#show

plt.figure(figsize=(40,20))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
